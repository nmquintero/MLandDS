### 2.1. Depth-First Search (DFS)

DFS is an uninformed search algorithm, part of the broader category of graph traversal techniques. It operates by visiting a node or state in the problem space and then recursively explores as far as possible along each branch before backtracking. Think of DFS as akin to exploring a maze by taking a path until you reach a dead-end, and then retracing your steps to explore other paths until you find the exit.

At its core, DFS relies on a stack data structure to maintain the nodes that need to be explored. When a node is visited, its adjacent nodes are pushed onto the stack for further exploration. The algorithm continues this process until all nodes are visited or the goal state is found.

Here's a step-by-step description of the Depth-First Search (DFS) algorithm:

`Step 1.` Initialize the Stack and Visited Set:
   - Create an empty stack to keep track of nodes to explore.
   - Create an empty set to store the nodes that have been visited.

`Step 2.` Push the Start Node onto the Stack and Mark as Visited:
   - Start the DFS by pushing the starting node (initial state) onto the stack.
   - Mark the starting node as visited by adding it to the visited set.

`Step 3.` Repeat Until the Stack is Empty:
   - While the stack is not empty, do the following steps:

`Step 4.` Pop a Node from the Stack:
   - Remove the top node from the stack. This node will be the current node that we are exploring.

`Step 5.` Process the Current Node:
   - Examine the current node to determine if it is the goal state or if it has any unvisited neighbors that need to be explored.

`Step 6.` Explore Unvisited Neighbors:
   - If the current node has unvisited neighbors, select one of them (based on a specific order, such as left-to-right or based on some heuristic if applicable).
   - Push the selected unvisited neighbor onto the stack.
   - Mark the neighbor as visited by adding it to the visited set.

`Step 7.` Backtrack if Necessary:
   - If the current node has no unvisited neighbors or is a dead-end (no more possible moves), backtrack by popping the previous node from the stack. This step is crucial in DFS as it allows the algorithm to explore other branches and paths.

`Step 8.` Repeat the Process:
   - Repeat `steps 4 to 7` until the stack becomes empty, which means all possible paths have been explored, or the goal state is found.

`Step 9.` Final Output:
   - If the goal state is found, the algorithm terminates, and the path or solution is obtained.
   - If the stack becomes empty without finding the goal state, then no solution exists for the given problem.

>It's important to note that the DFS algorithm explores a single branch in depth before backtracking, making it a suitable choice for solving problems with large state spaces and situations where finding any solution quickly is more important than finding the optimal solution.

_Advantages of Depth-First Search:_

1. Memory Efficiency: DFS has relatively low memory requirements as it explores a single branch in-depth before backtracking. It is well-suited for problems with large state spaces, where memory consumption is a concern.

2. Completeness (In Finite Spaces): In problems with finite state spaces, DFS is guaranteed to find a solution if one exists. This property holds because it explores all possible paths.

3. Applicability in Certain Scenarios: DFS is particularly useful in scenarios where finding any solution quickly is more important than finding the optimal solution. It is commonly applied in pathfinding and maze-solving scenarios.

_Limitations of Depth-First Search:_

1. Completeness (_In Infinite Spaces_): DFS may not be complete in infinite state spaces or graphs with loops. In such cases, it can get trapped in cycles and fail to find a solution even if one exists.

2. Lack of Optimal Solution: DFS does not guarantee finding the optimal solution. It might find a solution faster, but that solution may not be the most optimal one.

3. Memory Usage (_In Infinite Spaces_): In infinite state spaces or deeply branching problems, DFS may consume a lot of memory due to its depth-first exploration, potentially leading to stack overflow.

DFS can be used in various scenarios, including graph traversal, pathfinding, puzzle-solving, and more. However, it has its limitations, such as potentially getting stuck in infinite loops (infinite state spaces) or not guaranteeing an optimal solution.

By carefully implementing the steps mentioned above, the Depth-First Search algorithm provides a systematic approach to explore problem spaces effectively and find solutions in various AI applications.

Python Implementation of Depth-First Search:

```python
def dfs(graph, node, visited):
    if node not in visited:
        print("Visiting node:", node)
        visited.add(node)
        for neighbor in graph[node]:
            dfs(graph, neighbor, visited)

# Example Usage:
graph = {
    'A': ['B', 'C', 'E'],
    'B': ['D', 'F'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

visited_nodes = set()
dfs(graph, 'A', visited_nodes)
```

### 2.2. Breadth-First Search (BFS)

In the realm of Artificial Intelligence and algorithms, BFS is a powerful technique for traversing and exploring problem spaces in a systematic manner. Unlike Depth-First Search, BFS focuses on exploring all neighboring nodes at the current level before moving to deeper levels, resembling the ripples formed when a stone is dropped in water.

Breadth-First Search is an uninformed search algorithm that operates by systematically exploring the immediate neighbors of a given node before moving on to their respective neighbors. Think of BFS as exploring all the nearby possibilities before venturing into deeper levels. This approach ensures that BFS searches for the shortest path or the shallowest solution first.

BFS uses a queue data structure to maintain the nodes to be explored. The algorithm starts from the initial state (start node) and expands the search to all neighboring nodes at the same level, then moves to their neighbors in the subsequent level until the goal state is found or all possible states are explored.

Step-by-Step Algorithm for Breadth-First Search (BFS):

`Step 1.` Initialize the Queue and Visited Set:
   - Create an empty queue to store nodes to explore.
   - Create an empty set to keep track of visited nodes.

`Step 2.` Enqueue the Start Node and Mark as Visited:
   - Begin the BFS by enqueuing the starting node (initial state) into the queue.
   - Mark the starting node as visited by adding it to the visited set.

`Step 3.` Repeat Until the Queue is Empty:
   - While the queue is not empty, do the following steps:

`Step 4.` Dequeue a Node from the Queue:
   - Remove the front node from the queue. This node will be the current node that we are exploring.

`Step 5.` Process the Current Node:
   - Examine the current node to determine if it is the goal state or if it has any unvisited neighbors that need to be explored.

`Step 6.` Explore Unvisited Neighbors:
   - If the current node has unvisited neighbors, enqueue them into the queue.
   - Mark the neighbors as visited by adding them to the visited set.

`Step 7.` Repeat the Process:
   - Repeat `steps 4` to `6` until the queue becomes empty, which means all nodes at the current level and their neighbors have been explored.

`Step 8.` Final Output:
   - If the goal state is found, the algorithm terminates, and the path or solution is obtained.
   - If the queue becomes empty without finding the goal state, then no solution exists for the given problem.

Let's explore a simple Python implementation of BFS for a basic graph traversal scenario. Suppose we have a graph represented as an adjacency list:

```python
from collections import deque

def bfs(graph, start):
    queue = deque([start])
    visited = set([start])

    while queue:
        current_node = queue.popleft()
        print("Visiting node:", current_node)

        for neighbor in graph[current_node]:
            if neighbor not in visited:
                queue.append(neighbor)
                visited.add(neighbor)

# Example Usage:
graph = {
    'A': ['B', 'C', 'E'],
    'B': ['D', 'F'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

bfs(graph, 'A')
```

*Advantages of Breadth-First Search:*

1. Guaranteed Optimal Solution: BFS guarantees finding the optimal solution (shortest path) for problems with unit path costs, as it explores nodes level by level.

2. Completeness (In Finite Spaces): For problems with finite state spaces, BFS is complete, meaning it will find a solution if one exists.

3. Suitable for Pathfinding: BFS is particularly useful for finding the shortest path in unweighted graphs, making it ideal for pathfinding scenarios.

*Limitations of Breadth-First Search:*

1. Memory Usage: BFS may consume significant memory, especially in graphs with deep levels or an infinite state space.

2. Inefficiency (In Infinite Spaces): In infinite state spaces, BFS may take a long time to explore all nodes at a given level before moving to the next level.

### 2.4. Depth-Limited Search and Iterative Deepening

**Depth-Limited Search** is a variation of Depth-First Search (DFS) that imposes a maximum depth or level on the search, preventing it from exploring beyond a specified depth limit. It combines the advantages of DFS's memory efficiency with the completeness of BFS within a limited depth range. DLS is particularly useful in scenarios with limited memory resources or infinite state spaces, where traditional DFS may get stuck in infinite loops.

Step-by-Step Algorithm for Depth-Limited Search (DLS):

`Step 1.` Initialize the Stack and Visited Set:
   - Create an empty stack to store nodes to explore.
   - Create an empty set to keep track of visited nodes.

`Step 2.` Push the Start Node onto the Stack and Mark as Visited:
   - Start the DLS by pushing the starting node (initial state) onto the stack.
   - Mark the starting node as visited by adding it to the visited set.

`Step 3.` Set the Depth Limit:
   - Define a depth limit, denoted as "depth_limit," which specifies the maximum depth the search is allowed to explore.

`Step 4.` Repeat Until the Stack is Empty or Depth Limit is Reached:
   - While the stack is not empty and the depth limit has not been reached, do the following steps:

`Step 5.` Pop a Node from the Stack:
   - Remove the top node from the stack. This node will be the current node that we are exploring.

`Step 6.` Process the Current Node:
   - Examine the current node to determine if it is the goal state or if it has any unvisited neighbors within the depth limit that need to be explored.

`Step 7.` Explore Unvisited Neighbors within Depth Limit:
   - If the current node has unvisited neighbors at a depth less than or equal to the depth limit, select one of them.
   - Push the selected unvisited neighbor onto the stack.
   - Mark the neighbor as visited by adding it to the visited set.

`Step 8.` Repeat the Process:
   - Repeat `steps 5` to `7` until the stack becomes empty or the depth limit is reached.

`Step 9.` Final Output:
   - If the goal state is found, the algorithm terminates, and the path or solution is obtained.
   - If the depth limit is reached without finding the goal state, increment the depth limit and repeat `steps 2` to `8` until the goal is found or deemed unreachable within the expanded depth limits.

Below is an example of Depth-Limited Search (DLS) implemented in Python. In this example, we'll perform DLS on a simple binary tree to find a target value within a specified depth limit.

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def depth_limited_search(node, target, depth_limit):
    if node is None:
        return False

    if node.val == target:
        return True

    if depth_limit == 0:
        return False

    # Perform DLS on left and right subtrees with reduced depth limit
    return depth_limited_search(node.left, target, depth_limit - 1) or depth_limited_search(node.right, target, depth_limit - 1)

# Example Binary Tree:
#         1
#        / \
#       2   3
#      / \   \
#     4   5   6

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.right = TreeNode(6)

# DLS Example
target_value = 6
depth_limit = 2

result = depth_limited_search(root, target_value, depth_limit)

if result:
    print(f"Target value {target_value} found within depth limit {depth_limit}.")
else:
    print(f"Target value {target_value} not found within depth limit {depth_limit}.")
```

In this example, we define a `TreeNode` class to represent each node of the binary tree. The `depth_limited_search` function takes the current node, the target value to be found, and the depth limit as input parameters.

The function recursively explores the binary tree, checking if the current node's value matches the target value. If not, it performs DLS on the left and right subtrees with a reduced depth limit until the depth limit reaches 0 or the target value is found.

For the given binary tree with the depth limit of 2, the function searches for the target value 5. The output will indicate whether the target value was found within the specified depth limit or not.

Keep in mind that this example is a simplified demonstration of DLS on a binary tree. In practice, DLS can be applied to more complex problem spaces and search scenarios.

**Iterative Deepening** is an intelligent search strategy that uses Depth-Limited Search (DLS) with an incrementally increasing depth limit. It repeatedly applies DLS with increasing depth limits until the goal state is found or a specified maximum depth is reached. This approach combines the benefits of BFS's completeness and DFS's memory efficiency.

Step-by-Step Algorithm for Iterative Deepening:

`Step 1.` Set Initial Depth Limit:
   - Define an initial depth limit, denoted as "initial_depth_limit."

`Step 2.` Repeat Until Goal is Found or Max Depth is Reached:
   - While the goal state is not found and the current depth limit is less than or equal to the maximum depth limit, do the following steps:

`Step 3.` Perform Depth-Limited Search:
   - Apply Depth-Limited Search (DLS) with the current depth limit on the problem space.

`Step 4.` If Goal is Found, Terminate:
   - If the goal state is found during DLS, terminate the search and output the path or solution.

`Step 5.` Increment Depth Limit:
   - If the goal state is not found within the current depth limit, increment the depth limit.

`Step 6.` Repeat the Process:
   - Repeat `steps 2` to `5` until the goal state is found or the maximum depth limit is reached.

Iterative Deepening effectively combines the completeness of BFS with the memory efficiency of DFS. It allows the search to explore deeper levels progressively, avoiding the need to store a massive search tree in memory, making it particularly useful in scenarios with limited resources and infinite state spaces.

Below is an example of Iterative Deepening implemented in Python. In this example, we'll use Iterative Deepening to find a target value in a binary tree. The depth limit will be incrementally increased until the target value is found or a specified maximum depth is reached.


```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def depth_limited_search(node, target, depth_limit):
    if node is None:
        return False

    if node.val == target:
        return True

    if depth_limit == 0:
        return False

    # Perform DLS on left and right subtrees with reduced depth limit
    return depth_limited_search(node.left, target, depth_limit - 1) or depth_limited_search(node.right, target, depth_limit - 1)

def iterative_deepening(root, target, max_depth):
    for depth_limit in range(max_depth + 1):
        if depth_limited_search(root, target, depth_limit):
            return True
    return False

# Example Binary Tree:
#         1
#        / \
#       2   3
#      / \   \
#     4   5   6

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.right = TreeNode(6)

# Iterative Deepening Example
target_value = 5
max_depth = 3

result = iterative_deepening(root, target_value, max_depth)

if result:
    print(f"Target value {target_value} found within maximum depth of {max_depth}.")
else:
    print(f"Target value {target_value} not found within maximum depth of {max_depth}.")
```


In this example, we define a `TreeNode` class to represent each node of the binary tree. The `depth_limited_search` function is the same as the one used in the Depth-Limited Search example. The `iterative_deepening` function takes the root of the binary tree, the target value to be found, and the maximum depth as input parameters.

The function iteratively performs Depth-Limited Search (DLS) with increasing depth limits, starting from 0 up to the specified maximum depth. If the target value is found within any of the depth limits, the function returns True. Otherwise, if the maximum depth is reached without finding the target value, the function returns False.

For the given binary tree with a maximum depth of 3, the function searches for the target value 5. The output will indicate whether the target value was found within the specified maximum depth or not.

Iterative Deepening is a useful technique to combine the advantages of both Breadth-First Search (BFS) and Depth-First Search (DFS) by exploring deeper levels progressively while avoiding the need to store a massive search tree in memory.

### 2.5. Realisation of BFS and DFS in Graphs
Let's make some preparations: create a graph and import the necessary libraries. We'll also announce what we're going to search for and where we're going to find it:

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.drawing.nx_pydot import graphviz_layout

# Define the graph
G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4, 5, 6, 7])
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(2, 4)
G.add_edge(2, 5)
G.add_edge(3, 6)
G.add_edge(3, 7)


# Define the start and goal nodes
start_node = 1
goal_node = 2
```

Now let's implement the search in depth: to do this, declare the array of vertices we have already passed (to avoid repetition) and the stack on which we will move (exactly the stack, because we are moving in depth):

```python
# Define the depth-first search function
def dfs(node):
    print('DFS:')
    visited = []
    stack = [node]
    while stack:
        current_node = stack.pop()
        print('Node', current_node)
        if current_node == goal_node:
            visited.append(current_node)
            return list(visited)
        visited.append(current_node)
        for neighbor in G.neighbors(current_node):
            if neighbor not in visited:
                stack.append(neighbor)
    return None
```

Unlike the previous algorithm, the breadth search does not have a stack on which to move, but a queue to which we add vertices that will be considered later:

```python
# Define the breadth-first search function
def bfs(node):
    print('BFS:')
    visited = []
    queue = [[node]]
    while queue:
        path = queue.pop(0)
        current_node = path[-1]
        print('Node', current_node)
        if current_node == goal_node:
            visited.append(current_node)
            return list(path)
        visited.append(current_node)
        for neighbor in G.neighbors(current_node):
            if neighbor not in visited:
                new_path = path.copy()
                new_path.append(neighbor)
                queue.append(new_path)
    return None
```

Perform both of them:

```python
# Perform the depth-first search
dfs_path = dfs(start_node)
# Perform the breadth-first search
bfs_path = bfs(start_node)
```

### 2.6. Tower of Hanoi
It seems like a child's game, but in fact it is one of the classic tasks in programming. One of the ways to solve it can be a well-known search, in particular in width and depth.

The Tower of Hanoi is a classic mathematical puzzle or game that consists of three pegs and a number of disks of different sizes. The objective of the puzzle is to move all the disks from one peg (the source) to another peg (the destination), using the third peg (the auxiliary), while following a set of rules:

Only one disk can be moved at a time.
Each move consists of taking the top disk from one of the pegs and placing it on top of another peg or on an empty peg.
A larger disk cannot be placed on top of a smaller disk.

The puzzle starts with all the disks stacked on one peg in decreasing order of size, with the largest disk at the bottom and the smallest disk at the top. The goal is to move all the disks to another peg, maintaining the same order of sizes.

The minimum number of moves required to solve the Tower of Hanoi puzzle with `n` disks is `2^n - 1`.

```python
def hanoi_dfs(n):
    start_state = tuple([tuple(range(n, 0, -1)), (), ()])
    goal_state = ((), (), tuple(range(n, 0, -1)))

    stack = [(start_state, [])]
    visited = set([start_state])

    while stack:
        state, path = stack.pop()

        if state == goal_state:
            return path

        for move in legal_moves(state):
            new_towers = [list(tower) for tower in state]
            new_towers[move[1]].append(new_towers[move[0]].pop())

            new_state = tuple([tuple(tower) for tower in new_towers])
            if new_state not in visited:
                visited.add(new_state)
                stack.append((new_state, path + [move]))

    return None

def hanoi_bfs(n):
    start_state = tuple([tuple(range(n, 0, -1)), (), ()])
    goal_state = ((), (), tuple(range(n, 0, -1)))

    queue = [(start_state, [])]
    visited = set([start_state])

    while queue:
        state, path = queue.pop(0)

        if state == goal_state:
            return path

        for move in legal_moves(state):
            new_towers = [list(tower) for tower in state]
            new_towers[move[1]].append(new_towers[move[0]].pop())

            new_state = tuple([tuple(tower) for tower in new_towers])
            if new_state not in visited:
                visited.add(new_state)
                queue.append((new_state, path + [move]))

    return None

def legal_moves(state):
    moves = []
    for i in range(len(state)):
        for j in range(len(state)):
            if i != j and (state[i] and (not state[j] or state[i][-1] < state[j][-1])):
                moves.append((i, j))
    return moves

def print_solution(moves):
    if moves is None:
        print("No solution found.")
    else:
        for move in moves:
            print("Move disk from tower", move[0] + 1, "to tower", move[1] + 1)


n = 2  # Number of disks
print("DFS:")
dfs_moves = hanoi_dfs(n)
print_solution(dfs_moves)

print("\nBFS:")
bfs_moves = hanoi_bfs(n)
print_solution(bfs_moves)
```
